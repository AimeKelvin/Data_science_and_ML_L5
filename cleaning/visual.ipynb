```python
import pandas as pd
sales= pd.read_csv("messy_sales.csv")
sales.describe()
```

```python
#getting unique values
print("Region",sales['Region'].unique())
print("Category",sales['Category'].unique())
print("Product",sales['Product'].unique())
```

```python
# Checking null values (NaN)
sales.isnull().sum()
```

```python
#Checking duplicated values
sales.duplicated().sum()
```

```python
#Data preparation by converting a string numbers to numbers
sales['Units Sold']=pd.to_numeric(sales['Units Sold'], errors='coerce')
sales['Total Sales']=pd.to_numeric(sales['Total Sales'], errors='coerce')
sales.info()
```

```python
print(sales['Units Sold'])
```

```python
#filling missing numeric values
from sklearn.impute import KNNImputer, SimpleImputer


s_imputer = SimpleImputer()#Average
k_imputer = KNNImputer()# Average for 5 elements
new_sales= sales.copy()

new_sales['Units Sold']= s_imputer.fit_transform(new_sales[["Units Sold"]])
new_sales['Units Sold']= k_imputer.fit_transform(new_sales[["Units Sold"]])

new_sales.isna().sum()
```

```python
# filling missing numeric values using statistical method
sales1= sales.copy()
sales2= sales.copy()
sales3= sales.copy()

mean= sales1['Units Sold'].mean()
median= sales1['Total Sales'].median()
sales1['Units Sold']= sales1['Units Sold'].fillna(mean)
sales1['Total Sales']= sales1['Total Sales'].fillna(median)
sales1.info()
```

```python
# fill missing numeric values using constant method
sales2['Units Sold']= sales2['Units Sold'].fillna(0)
sales2['Total Sales']= sales2['Total Sales'].fillna(1000)

sales2
```

```python
#Deleting missing values
sales3.dropna(inplace=True)## if you want the current Data set to change use "inplace=True"
sales3.info()
```

```python
#Standardize text columns( make names consistent)
sales['Category'] = sales['Category'].str.lower()
sales['Category'] = sales['Category'].str.replace('gadgets','gadget')

sales['Category'].unique()
sales.info()
```

```python
sales['Unit Price']= sales['Unit Price'].str.replace('$','').str.strip() ## strip is for removind exctra spaces
sales['Unit Price']
```
