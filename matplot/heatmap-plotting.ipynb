```python
import matplotlib.pyplot as pt
import numpy as np

scores = [23, 45, 56, 78, 213, 45, 67, 89, 90, 123, 0, 167, 189, 210, 23, 200]

# Convert to 2D (1 row, 16 columns)
data = np.array(scores).reshape(1, -1)

pt.imshow(data, cmap='viridis', aspect='auto')
pt.colorbar(label='Scores')
pt.xlabel("Hour Index")
pt.ylabel("Intensity")
pt.title("Scores Heatmap", fontsize=14, fontweight='bold')
pt.show()
```
