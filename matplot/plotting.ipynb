```python
import matplotlib.pyplot as plt

scores = [23, 45, 56, 78, 213, 45, 67, 89, 90, 123, 0, 167, 189, 210, 23, 200]

# Line Chart
plt.figure()
plt.plot(scores, marker='o', label="Scores")
plt.xlabel("Index")
plt.ylabel("Scores")
plt.title("Line Chart of Scores")
plt.legend()
plt.grid(True)
plt.show()

# Bar Chart
plt.figure()
plt.bar(range(len(scores)), scores, label="Scores")
plt.xlabel("Index")
plt.ylabel("Scores")
plt.title("Bar Chart of Scores")
plt.legend()
plt.grid(True)
plt.show()

# Scatter Chart
plt.figure()
plt.scatter(range(len(scores)), scores, label="Scores")
plt.xlabel("Index")
plt.ylabel("Scores")
plt.title("Scatter Chart of Scores")
plt.legend()
plt.grid(True)
plt.show()
```
