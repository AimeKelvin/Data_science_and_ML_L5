```python
#TASK 2
df = pd.read_csv('/mnt/data/student_performance.csv')
```

```python
#TASK 3


# Check for missing values
df.isnull().sum()

# Check for duplicates
df[df.duplicated()]

# Check for invalid values (example for numerical columns)
for col in ['math score', 'reading score', 'writing score']:
    print(f"{col} - min: {df[col].min()}, max: {df[col].max()}")

# Check for categorical anomalies
for col in ['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course']:
    print(f"{col} unique values: {df[col].unique()}")

```

```python
#TASK 4

# 1. Remove duplicates
df = df.drop_duplicates()

# 2. Fill missing numerical values (example: fill with mean)
for col in ['math score', 'reading score', 'writing score']:
    df[col].fillna(df[col].mean(), inplace=True)

# 3. Fill missing categorical values (example: mode)
for col in ['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course']:
    df[col].fillna(df[col].mode()[0], inplace=True)

# 4. Correct data types if needed
# Example: convert scores to integer
for col in ['math score', 'reading score', 'writing score']:
    df[col] = df[col].astype(int)

# Check cleaned dataset
df.info()
df.head()
```
